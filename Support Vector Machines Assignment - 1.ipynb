# Q1. What is the mathematical formula for a linear SVM?

The mathematical formula for a linear SVM is given by:

$$f(x) = \textbf{w}^T\textbf{x} + b$$

where $\textbf{x}$ is the input vector, $\textbf{w}$ is the weight vector, and $b$ is the bias term. The goal of the SVM algorithm is to find the optimal hyperplane that separates the data points in different classes in the feature space. The hyperplane is defined by the equation $f(x) = 0$, which corresponds to the decision boundary of the SVM. The weight vector $\textbf{w}$ is perpendicular to the hyperplane and determines the orientation of the decision boundary, while the bias term $b$ shifts the decision boundary along the direction of the weight vector.

# Q2. What is the objective function of a linear SVM?

The objective function of a linear SVM is to find the optimal hyperplane in an N-dimensional space that can separate the data points in different classes in the feature space. The hyperplane tries that the margin between the closest points of different classes should be as maximum as possible ¹. The SVM algorithm tries to minimize the classification error while maximizing the margin between the decision boundary and the training data points. The optimization problem can be formulated as a quadratic programming problem, which involves minimizing the norm of the weight vector subject to the constraint that all training samples are correctly classified ³.

The objective function of a linear SVM can be written as:

$$\min_{\textbf{w},b} \frac{1}{2} \|\textbf{w}\|^2$$

subject to:

$$y_i(\textbf{w}^T\textbf{x}_i + b) \geq 1, \forall i = 1,2,...,n$$

where $\textbf{x}_i$ is the $i^{th}$ training sample, $y_i$ is its corresponding class label, and $n$ is the number of training samples. The optimization problem tries to find the weight vector $\textbf{w}$ and the bias term $b$ that satisfy the above constraints and minimize the norm of the weight vector ¹.

# Q3. What is the kernel trick in SVM?

The **kernel trick** is a powerful technique that enables SVMs to solve non-linear classification problems by implicitly mapping the input data to a higher-dimensional feature space. By doing so, it allows us to find a hyperplane that separates the different classes of data ¹. The kernel trick is used to transform the input data into a higher-dimensional space, where it becomes easier to separate the data points using a linear classifier. The kernel function computes the dot product between the transformed feature vectors in the higher-dimensional space, without actually computing the transformation explicitly ³. This makes it computationally efficient to work with high-dimensional feature spaces.

There are different types of kernel functions such as **linear**, **polynomial**, and **radial basis function (RBF)**. The choice of kernel function depends on the specific problem and the characteristics of the data. For instance, the linear kernel is used for linearly separable data, while the RBF kernel is used for non-linearly separable data ².

# Q6. SVM Implementation through Iris dataset.

- Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

## Load the iris dataset
iris = load_iris()

# Split the dataset into a training set and a testing set
X_train , X_test , Y_train , Y_test = train_test_split(iris.data,iris.target,random_state=42,test_size=0.3)
```

- Train a linear SVM classifier on the training set and predict the labels for the testing setl

```python
from sklearn.svm import LinearSVC

# Train a linear SVM classifier on the training set
clf = LinearSVC(random_state=42)
clf.fit(X_train,Y_train)

# Predict the labels for the testing set
y_pred = clf.predict(X_test)
```

- Compute the accuracy of the model on the testing setl

```python
from sklearn.metrics import accuracy_score

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(Y_test,y_pred)
accuracy
```

- Plot the decision boundaries of the trained model using two of the featuresl

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.tree import DecisionTreeClassifier 

# Choosing the first 2 columns for the plot
X_train_cols = X_train[:, :2] 

# Creating and fitting the tree classifier
classifier = DecisionTreeClassifier(max_depth=4, 
                                    random_state=42).fit(X_train_cols, Y_train)

# Plotting the tree boundaries
disp = DecisionBoundaryDisplay.from_estimator(classifier, 
                                              X_train_cols, 
                                              response_method="predict",
                                              xlabel=iris.feature_names[0], ylabel=iris.feature_names[1],
                                              alpha=0.5, 
                                              cmap=plt.cm.coolwarm)

# Plotting the data points    
disp.ax_.scatter(X_train[:, 0], X_train[:, 1], 
                 c=Y_train, edgecolor="k",
                 cmap=plt.cm.coolwarm)

plt.title(f"Decision surface for tree trained on {iris.feature_names[0]} and {iris.feature_names[1]}")
plt.show()
```

- Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = load_iris()

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Try different values of C
for C in [0.1, 1, 10, 100]:
    # Train a linear SVM classifier on the training set
    clf = LinearSVC(C=C, random_state=42)
    clf.fit(X_train, y_train)

    # Predict the labels for the testing set
    y_pred = clf.predict(X_test)

    # Compute the accuracy of the model on the testing set
    accuracy = accuracy_score(y_test, y_pred)

    print(f"The accuracy of the model on the testing set with C={C} is {accuracy:.2f}")
```
