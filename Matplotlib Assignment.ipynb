Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

* Note: Also add title, xlabel, and ylabel to the plot.

* Scatter plot is one of the function to print the graph of a data . This is used to print graph in the form of dot or scatter we can add title , xlable , ylabel , colour ,side etc in scatter graph . This is one of the best graph function in pyplot . We can understand it by some example 

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('This is scatter graph')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

* import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

* The subplot() function is used to create multiple plots in one figure. It helps in creating common layouts of subplots and it also includes the enclosing figure object, in a single call123.

Here’s an example of how to draw four line plots using the subplot() function:

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x,y1)

plt.subplot(2, 2, 2)
plt.plot(x,y2)

plt.subplot(2, 2, 3)
plt.plot(x,y3)

plt.subplot(2, 2, 4)
plt.plot(x,y4)

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

* Bar plot is one of the graph used to plot the data of str and int together . In this graph we have 2 types of data first is str means some name and second is float or an int we used plt.bar to print graph of this type of data. we can plot this data in 2 form horizontally and vertically . Vertically the str is in the x axis and int numbers is in y axis  and horizontally str is in y axis and int is is x axis 

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Vertical bar plot

plt.bar(company,profit)
```

```python
# Horizontally bar plot
plt.barh(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

* A box plot is a type of chart that depicts a group of numerical data through their quartiles. It is a simple way to visualize the shape of our data. It makes comparing characteristics of data between categories very easy. Box plots divide the data into sections containing approximately 25% of the data in that set. Box plots are useful as they provide a visual summary of the data enabling researchers to quickly identify mean values, the dispersion of the data set, and signs of skewness123.

Here’s an example of how to draw a box plot using the data you provided:

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig = plt.figure(figsize =(10, 7))

ax = fig.add_subplot(111)

bp = ax.boxplot(data)

plt.show()
```
